```python
def convert_minutes(x):
    min = list(range(5,56, 5))
    for m in min:
        if x % m == x and x > m-5:
            return m
        if x in [56,57,58,59]:
            return 0
        if x in min+[0]:
            return x
```

```python
convert_minutes(4)
```

```python
df['minute'] = df['minute'].apply(lambda x: convert_minutes(x))
```

```python
plt.figure(figsize=(5,7))
sns.countplot(data=df, y='minute')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title('Minute', fontsize=14)
plt.xlabel('')
plt.ylabel('')
plt.tight_layout()
```

 - Most of the accidents happened between 2pm to 7pm and it was minimum between midnight till 6'o clock in the morning
 

 - Most of the fatal accidents happened on weekends.
 
 
 - Drivers with age more than 18 cause more serious injuries as compared to those below 18
 
 
 - Maximum number of serious and fatal injuries happened to junior high school
 
 
 - Drivers with driving experience with 5-10 years met with proportionately less number of serious injuries as compared to
  other age groups
  
  
 - Maximum no of vehicle met with accedent falls under type: automobile
 
 
 - Maximum cases of accedents happened to self_owned vehicles with no defects
 
 
 - Though most of the accident happened around offices, a higher proportion of accidents happened around residential areas have led to fatal injuries.
 
 
 - Maximum accidents happened on tangent road with flat terrain, the road surface type was asphalt road
 
 
 - The junction type met with maximum accidents was Y-shaped junction
 
 
 - Maximum accidents happened in daylight which involed two vehicles and the type of collision was vehicle to vehicle colision in normal weather conditions
 
 
 - In most of the cases number of casuality was one
 
 
 - Most of the cases of casualty happened to drivers as compared to pedestrians or passengers
 
 
 - The conditions on which most of the drivers met with the accident was straight movement
 
 
 - Maximum number of fatal injures happened at night
 
 
 - Cause of accidents in most of the cases were : no distancing, changing lane to the left or right and driving carelessly
 
 
 -  The lanes undivided two way and Two way (divided with broken line road markings) met with maximum occurances of accidents

# Preprocessing

```python
df.isnull().sum()
```

### Dropping the columns with high imbalance as they can cause imbalance during imputation

```python
cols = ['Defect_of_vehicle', 'Vehicle_driver_relation', 'Work_of_casuality', 'Fitness_of_casuality'] 
```

```python
for i in cols:
    sns.countplot(y=i,data=df)
    plt.yticks(fontsize=14)
    plt.ylabel('')
    plt.xlabel('')
    plt.title(i,fontsize=14)
    plt.show()
```

```python
df.drop(['Defect_of_vehicle', 'Vehicle_driver_relation', 'Work_of_casuality', 'Fitness_of_casuality'],axis=1,inplace=True)
```

```python
df.shape
```

### Imputing the null values

```python
imput_cols = df.isna().sum()[df.isna().sum() != 0].index.tolist()
```

```python
imput_cols
```

```python
for i in imput_cols:
    mode = df[i].mode()[0]
    df[i].fillna(mode,inplace=True)
```

```python
df.isnull().sum()
```

```python
cat_cols = df.select_dtypes(include="O")
```

```python
cat_cols
```

```python
cols = cat_cols.columns
cols
```

```python
for i in cols:
    if cat_cols[cat_cols[i]=='na'].empty == False:
        print(i)
```

```python
for i in ['Casualty_class','Sex_of_casualty','Age_band_of_casualty','Casualty_severity']:
    cat_cols[i] = cat_cols[i].replace('na',cat_cols[i].mode()[0])
```

```python
for i in cols:
    if cat_cols[cat_cols[i]=='na'].empty == False:
        print(i)
```

```python
cat_cols['Age_band_of_casualty'].value_counts().index[1]
```

```python
cat_cols['Age_band_of_casualty'] = cat_cols['Age_band_of_casualty'].replace('na',cat_cols['Age_band_of_casualty'].value_counts().index[1])
```

```python
for i in cols:
    if cat_cols[cat_cols[i]=='na'].empty == False:
        print(i)
```

```python
cat_cols
```

```python
cols = ['Day_of_week', 'Age_band_of_driver', 'Sex_of_driver',
       'Educational_level', 'Driving_experience', 'Type_of_vehicle',
       'Owner_of_vehicle', 'Service_year_of_vehicle', 'Area_accident_occured',
       'Lanes_or_Medians', 'Road_allignment', 'Types_of_Junction',
       'Road_surface_type', 'Road_surface_conditions', 'Light_conditions',
       'Weather_conditions', 'Type_of_collision', 'Vehicle_movement',
       'Casualty_class', 'Sex_of_casualty', 'Age_band_of_casualty',
       'Casualty_severity', 'Pedestrian_movement', 'Cause_of_accident',
       'Number_of_vehicles_involved','Number_of_casualties', 'hour', 'minute','Accident_severity']
```

```python
final_df = final_df[cols]
```

```python
final_df
```

```python
X = final_df.iloc[:,:1].values
y = final_df.iloc[:,-1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, shuffle=True)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

### Upsampling
